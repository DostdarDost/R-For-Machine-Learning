### While loop
 while loop is the simplest loop, which executes one or more statements, if the given condition remains true. it is useful when the number of iterations is not known in advance.

```python
# Example
i=1
while i<6:
  print(i)
  i +=1
```

### For loop
A for loop is used for iterating over a sequence (that is either a list, atuple, a dictoinary. a set or string).

This is less like the for keyword in other programming languages, and works more like an iterator method as found in other object-orientated programming languages.

With the for loop we can execute a set of statements, once for each item in a list, tuple, set etc.


```python
# Example:
languages=['swift', 'Python','Go','JavaScript']
for language in languages:
  print(language)
```
