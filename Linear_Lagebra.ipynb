### 2. Numerical Linear Algebra
According to Wikipedia the numerical linear algebra is define as:"Numerical linear algebra, sometimes called applied linear algebra, is the study of how matrix operations can be used to create computer algorithms which efficiently and accurately provide approximate answers to questions in continuous mathematics. It is a subfield of numerical analysis, and a type of linear algebra. Computers use floating-point arithmetic and cannot exactly represent irrational data, so when a computer algorithm is applied to a matrix of data, it can sometimes increase the difference between a number stored in the computer and the true number that it is an approximation of. Numerical linear algebra uses properties of vectors and matrices to develop computer algorithms that minimize the error introduced by the computer, and is also concerned with ensuring that the algorithm is as efficient as possible"

### Combining Arrays

```python
# Create array with vstack
from numpy import array
from numpy import vstack
# Create first array 
a = array ([1,2,3])
print(a)
# create second array
b= array([4,5,6])
print(b)
# create third array
c= vstack([a,b])
print(c)
# size of the array
print(c.shape)
```

####  Slicing of NumPy Array
In meachine Learning slicing is very important to split data set into input variables and output variables or train dataset and test dataset. Slicing is specified using tbe colon operator : as  $data[from : to]$

```python
# slicing of one-demensional array
from numpy import array
# define array
data = array([11,33,44, 55, 66])
#display data
print(data[:])
```

The first item strat from index 0 end at 1 is slice like that

```python
# slice a subset of a one-dimensional array
from numpy import array 
# define array
data = array([11, 22, 33,44,55])
# slicing
print(data[0:1])
print(data[0:2])
print(data[0:3])
print(data[0:4])
print(data[0:5])
```

I will show you negative indexes in slices, let we need last two items 

```python
# negative slicing of a 1-d array
from numpy import array
#define data
data = array([11,22,33,44,55])
# negative slicing
print(data[-1:])
print(data[-2:])
print(data[-3:])
print(data[-4:])
print(data[-5:])
print(data[-5:])
print(data[-4:])
print(data[-3:])
print(data[-2:])
print(data[-1:])
```

Two-Dimensoinal Slicing is most likely to use in machine learning.


 **split input and output Features**

---



Real data set have input and output variables like X and y respectily. we need to split using slicing techniques all rows and all columns. let we need all columns except the last one.
 `X =[:,:-1]  y =[:,-1]`


---









```python
# split input and output data
from numpy import array
# define array
data = array([
    [11, 22, 33],
    [44, 55, 66],
    [77, 88, 99]])
# separate data
X,y =data[:,:-1],data[:,-1]
print(X)
print(y)
```

**Split train and Test Rows** 

---

It is commond in Maching Learning model traning, the real dataset selpit in to two or more more set according to requaiedment but train and test must be done durring train and test model.
 ` train = data[:split,:]   , test = data[split:,:]`

```python
# split train and test data
from numpy import array
 # define array
data = array([
    [11, 22, 33],
    [44, 55, 66],
    [77, 88, 99]])
#separate data
split= 2
train,test = data[:split , :],data[split: , :]
#display
print(train)
print(test)
```
